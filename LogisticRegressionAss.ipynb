```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
data = pd.read_csv("/content/bank-fullkaggle.csv")
data.head()
```

```python
data.info()
    
```

```python
data1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data1.head()
```

```python
data1.info()
```

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['Target'] = np.where(data1['Target'].str.contains("yes"), 1, 0)
data1
```

# **Preparing a Model**

```python
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

# **Model Prediction**

```python
y_pred = classifier.predict(x)
```

```python
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual': y,
                          'predicted_prob':classifier.predict(x)})
y_pred_df
```

```python
# Confusion matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
print (confusion_matrix)
```

# **ROC Curve**

```python
fpr,tpr, thresholds = roc_curve(y , classifier.predict_proba (x)[:,1])

auc = roc_auc_score(y,y_pred)

plt.plot(fpr , tpr , color='red', label='logit model (area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--') #k-- is diagonal
plt.xlabel('False Positive Rate or [1 - True Negative rate]')
plt.ylabel('True Positive Rate')
```
